```python
from tqdm import tqdm
from libs.feature_extraction.vectorizers import BertVectorizer
import pandas as pd
import numpy as np
bert = BertVectorizer("microsoft/unixcoder-base")
vectors = bert.load_vectors("data/comment_finder/vectors/bert_vectors.npy")  
df = pd.read_csv("./data/comment_finder/cosine_bleu_C10__classified.csv")
all_series = df.index
display(df.head())

for f in (1,2,3,4,5):
    test_df = pd.read_csv("./results/bert_cosine_bleu_C1_F5/fold_"+str(f)+".csv")

    train_data = pd.DataFrame()
    test_data = pd.DataFrame()

    train_vectors = []
    test_vectors = []

    for idx in tqdm(all_series):
        if test_df.shape[0]>0 and idx == test_df.iloc[0].test_idx:
            test_data= pd.concat([test_data, df.iloc[[idx]]], ignore_index=False)
            test_vectors.append(vectors[idx])
            test_df = test_df.drop(labels=0, axis=0)
            test_df = test_df.reset_index(drop=True)
        else:
            train_data = pd.concat([train_data, df.iloc[[idx]]], ignore_index=False)
            train_vectors.append(vectors[idx])
            
    train_data.rename(columns={'index':'test_idx'}, inplace=True)
    train_data = train_data.reset_index()
    
    test_data.rename(columns={'index':'test_idx'}, inplace=True)
    test_data = test_data.reset_index()
    
    display("Fold"+str(f))        
    display(df.shape)
    display(train_data.shape)
    display(test_data.shape)
    display(len(vectors))
    display(train_data.head())
    display(test_data.head())

    display(len(train_vectors))
    display(len(test_vectors))
    
    train_data.to_csv("train-test-data/fold"+str(f)+"/train_data.csv")
    test_data.to_csv("train-test-data/fold"+str(f)+"/test_data.csv")
    np.save("train-test-data/fold"+str(f)+"/train_vectors.npy", train_vectors)
    np.save("train-test-data/fold"+str(f)+"/test_vectors.npy", test_vectors)
    
```
